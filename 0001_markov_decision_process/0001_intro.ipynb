#### Markov Decision Process

Elements common to all control tasks

States ($S_t$): The possible conditions or states that an agent can be in 

Actions ($A_t$): The actions that an agent can take in each state 

Rewards ($R_t$): The numerical value assigned to each action taken in a state 

Agent: The entity that interacts with the environment and makes decisions 

Environment: The conditions that the agent interacts with 


<font color="skyblue">Discrete-time</font> <font color="green">stochastic</font> <font color="orange">control process</font>

<font color="skyblue">Discrete-time</font> : Time moves forward in finite intervals: $ t \text{ } \epsilon$ {1,2,3,...}.

<font color="green">stochastic</font> : Future states depend only partial on the actions taken.

<font color="orange">control process</font>: It is based on decision making to reach the target state.

#### $$( S, A, R, P)$$

1. Set of possible states of the task.
2. Set of actions that can be taken in each of the state.
3. Set of rewards for each (s, a) pair.
4. Probabilities of passing from one state to another when taking each possible action.

The process has no memory, the next state depends only on the current state, not on the previous once.

If a process meets this property, it is known as a Markov process.

The difference between Markov Chain & Markov Process are "Actions" & "Rewards"

Types of Markov decision process

1. Finite (Action, Reward, State, Environment - All have some limits - Simple Maze)
2. Infinite (Here there is no limit, Self Driving Car)

1. Episodic (Playing Chess would end at check mate)
2. Continuing (Does not have termination condition)

In this course Episodic ones will be covered

<b> Trajectory and Episode </b>

Trajectory: Elements that are generated when the agent moves from one state to another

$\tau$ = $S_0$,$A_0$,$R_1$,$S_1$,$A_1$,$R_2$,$S_2$,$A_2$,$R_3$,$S_3$,$A_3$

Episode: Trajectory from the initial state of the task to a terminal state.

$\tau$ = $S_0$,$A_0$,$R_1$,$S_1$,$A_1$,$R_2$,$S_2$,$A_2$,.....,$R_T$,$S_T$

<b> Reward Vs. Return </b>

The goals of the task are represented by the <b>rewards</b> $ R_t $.
1. We want to maximize the sum of rewards.
2. A short-term reward can worsen long-term results.

So we want to maximize the long-term sum of rewards <b> or </b> maximize the episode's return

Reward: $R_1$

Return: $G_t$ = $R_{t+1}$ + $R_{t+2}$ + $R_{t+3}$ + ..... + $R_t$
